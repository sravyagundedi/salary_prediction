```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 150],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 4],
    'subsample': [0.8, 1.0]
}

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=GradientBoostingClassifier(random_state=42),
    param_grid=param_grid,
    scoring='accuracy',
    cv=3,
    n_jobs=-1,
    verbose=1
)
```

```python

grid_search.fit(xtrain,ytrain)

best_model = grid_search.best_estimator_

y_pred = best_model.predict(xtest)
```

```python
for i in range(len(xtest)):
    print(f"Actual: {ytest.iloc[i]}, Predicted: {y_pred[i]}")
```

```python
accuracy_score(y_pred, ytest)
```

```python
import joblib
joblib.dump(best_model, "best_model.pkl")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# After fitting GridSearchCV
results = pd.DataFrame(grid_search.cv_results_)

# Find the best combination of n_estimators and subsample
best_n_estimators = grid_search.best_params_['n_estimators']
best_subsample = grid_search.best_params_['subsample']

# Filter results for the best n_estimators and subsample
filtered_results = results[(results['param_n_estimators'] == best_n_estimators) &
                           (results['param_subsample'] == best_subsample)]

# Create the pivot table
pivot_table = filtered_results.pivot(index='param_max_depth', columns='param_learning_rate', values='mean_test_score')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".3f", cmap="YlGnBu")
plt.title("GridSearchCV Mean Test Score (Best n_estimators and subsample)")
plt.xlabel("Learning Rate")
plt.ylabel("Max Depth")
plt.show()
```
